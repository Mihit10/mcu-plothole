```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import time

# Setup headless browser
options = Options()
options.add_argument("--headless")
driver = webdriver.Chrome(options=options)

url = "https://marvelcinematicuniverse.fandom.com/wiki/Ant-Man_and_the_Wasp"
driver.get(url)
time.sleep(3)  # Let content load

soup = BeautifulSoup(driver.page_source, "html.parser")
driver.quit()

# Function to extract content between headings
def extract_section(soup, section_id):
    section = soup.find("span", {"class": "mw-headline", "id": section_id})
    if not section:
        return f"Section '{section_id}' not found."

    content = []
    for sibling in section.parent.find_next_siblings():
        if sibling.name and sibling.name.startswith("h"):  # Stop at the next heading
            break
        content.append(sibling.get_text(strip=True))
    
    return "\n".join(content)

# Get sections
synopsis = extract_section(soup, "Synopsis")
plot = extract_section(soup, "Plot")

# Print or save
print("=== SYNOPSIS ===")
print(synopsis)

print("\n=== PLOT ===")
print(plot)
```
