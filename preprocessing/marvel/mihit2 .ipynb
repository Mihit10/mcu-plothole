```python
import spacy

# Load a smaller spaCy model (faster and less resource-intensive)
nlp = spacy.load("en_core_web_sm")

def extract_triplets_pattern(text):
    """
    Extracts subject-predicate-object triplets from the given text using dependency parsing
    and pattern-based matching.
    """
    doc = nlp(text)
    triplets = []

    for sent in doc.sents:
        for token in sent:
            # Look for verbs that connect entities
            if token.pos_ == "VERB":
                subject = None
                object_ = None

                # Find the subject (noun/pronoun to the left)
                for child in token.children:
                    if child.dep_ in ("nsubj", "nsubjpass"):
                        subject = child.text
                        break

                # Find the object (noun/pronoun to the right)
                for child in token.children:
                    if child.dep_ in ("dobj", "pobj"):
                        object_ = child.text
                        break

                # If both subject and object are found, create a triplet
                if subject and object_:
                    triplets.append({
                        "subject": subject,
                        "predicate": token.lemma_,  # Use lemma for the base form of the verb
                        "object": object_
                    })

    return triplets

# Example Usage
text = resolved_text
triplets = extract_triplets_pattern(text)
print(triplets)
```
